```python
import manim as mn
import random
from manim import *

config.media_width = "75%"
config.verbosity = "WARNING"

print(mn.__version__)
```

```python
%%manim -qm Splinkerette

# https://docs.manim.community/en/stable/reference/manim.utils.color.manim_colors.html

do_full_anim = True

#####################################
##################################### Splinkerette inverse PCR protocol
#####################################

dna_dy = 0.05

def make_dsDNA(x1,x2,y):
    o1 = Line(
        [x1,dna_dy,0], 
        [x2,dna_dy,0]
    )
    o2 = Line(
        [x1, -dna_dy,0], 
        [x2, -dna_dy,0]
    )    
    return [o1, o2]



def add_vec(u,v):
    return [
        u[0]+v[0],
        u[1]+v[1],
        u[2]+v[2]
    ]

class Splinkerette(MovingCameraScene):
    def construct(self):

        ######## constants #############

        
        def make_splinkerette(do_flip=1):
            dna_len = 0.2
            dna_len2 = 1
            
            o1 = Line(
                [do_flip*0,      +do_flip*dna_dy,0], 
                [do_flip*dna_len,+do_flip*dna_dy,0]
            )
            b = CurvedArrow(
                [do_flip*dna_len,+do_flip*dna_dy,0], 
                [do_flip*dna_len,-do_flip*dna_dy + do_flip*1.2,0], 
                radius= 0.6
            )
            o2 = Line(
                [do_flip*0,       -do_flip*dna_dy,0], 
                [do_flip*dna_len2,-do_flip*dna_dy,0]
            )
            return [o1,o2,b]
        

        dna_noins = make_dsDNA(0,4,0)
        self.add(*dna_noins)

        dna_left = VGroup(*make_dsDNA(-8,-4,0))
        self.add(dna_left)
        
        dna_right = VGroup(*make_dsDNA(4,8,0))
        self.add(*dna_right)

        dna_ins = make_dsDNA(-4,0,0)
        self.add(*dna_ins)        

        r = Rectangle(fill_color="RED", width=2, height=0.1, fill_opacity=1)
        r.shift([-2,0,0])
        self.add(r)

        group_ins = VGroup([r]+dna_ins)
        group_noins = VGroup(*dna_noins)



        ############# Show text ########################
        txt_intro1 = Tex(r"We want to amplify this \\ red insert and surrounding DNA")
        txt_intro1.move_to([-2,+1.5,0])        
        self.play(
            Create(txt_intro1)
        )
        self.wait(2)
        
        txt_intro2 = Tex(r"But we cannot construct a primer here! \\ (Sequence unknown)")
        txt_intro2.move_to([+2,-1.5,0])        
        self.play(
            Create(txt_intro2)
        )
        self.wait(2)


        txt_intro3 = Tex(r"This problem can be overcome by inverse PCR!")
        txt_intro3.move_to([0,-3,0])
        self.play(
            Create(txt_intro3)
        )
        self.wait(4)

        self.play(
            txt_intro1.animate.shift([0,8,0]),
            txt_intro2.animate.shift([0,8,0]),
            txt_intro3.animate.shift([0,-8,0])
        )
        


        
        ############# Show text ########################
        txt_fragment = Tex(r"The DNA is first fragmented (e.g. enzymatically)")
        txt_fragment.move_to([0,3,0])
        
        self.play(
            Create(txt_fragment)
        )
        self.wait(2)
                           
        
        ############# Perform fragmentation ########################
        
        pos_ins       = [-3.5,0,0]
        pos_noins     = [+3.5,0,0]
        pos_dna_left  = [-8,+2,0]
        pos_dna_right = [+8,+2,0]
        
        self.play(
            dna_left.animate.move_to(pos_dna_left),
            dna_right.animate.move_to(pos_dna_right),
            group_ins.animate.move_to(pos_ins),
            group_noins.animate.move_to(pos_noins)
        )
        self.wait(2)

        ############# Create text ########################
        
        self.play(
            txt_fragment.animate.shift([0,5,0])
        )
        txt_ligate = Tex(r"Splinkerette adapters are ligated")
        txt_ligate.move_to([0,3,0])
        self.play(
            Create(txt_ligate)
        )
        self.wait(2)

        ############# Perform ligation ########################

        move_lig_dy = 10
        spr_ins = make_splinkerette(1)
        spl_ins = make_splinkerette(-1)
        sr_ins = VGroup(*spr_ins)
        sl_ins = VGroup(*spl_ins)
        sr_ins.shift([0,-move_lig_dy,0])
        sl_ins.shift([0,-move_lig_dy,0])
        
        sr_noins = VGroup(*make_splinkerette(1))
        sl_noins = VGroup(*make_splinkerette(-1))
        sr_noins.shift([0,-move_lig_dy,0])
        sl_noins.shift([0,-move_lig_dy,0])
        
        sr_above = VGroup(*make_splinkerette(-1))
        sl_above = VGroup(*make_splinkerette(1))
        sr_above.shift([0,-move_lig_dy,0])
        sl_above.shift([0,-move_lig_dy,0])

        self.play(
            sl_ins.animate.shift(add_vec(pos_ins, [-2,move_lig_dy,0])),
            sr_ins.animate.shift(add_vec(pos_ins, [+2,move_lig_dy,0])),
            sl_noins.animate.shift(add_vec(pos_noins, [-2,move_lig_dy,0])),
            sr_noins.animate.shift(add_vec(pos_noins, [+2,move_lig_dy,0])),
            sl_above.animate.shift(add_vec(pos_dna_left, [+2,move_lig_dy,0])),
            sr_above.animate.shift(add_vec(pos_dna_right, [-2,move_lig_dy,0]))
        )        
        self.wait()


        ############# Create text ########################
        
        self.play(
            txt_ligate.animate.shift([0,5,0])
        )
        txt_pcr = Tex(r"Perform first cycle \\ PCR against insert")
        txt_pcr.move_to([-2,3,0])
        txt_pcr_empty = Tex("\"Empty\" DNA is not targeted")
        txt_pcr_empty.move_to([3,-2,0])
        self.play(
            Create(txt_pcr)
        )
        self.wait(1)
        self.play(
            Create(txt_pcr_empty)
        )
        self.wait(1)
        
        ############# First round PCR - "heat up" ########################
        
        group_upper_strand = VGroup([dna_ins[0]], spr_ins[0], spr_ins[2], spl_ins[1])
        self.play(
            group_upper_strand.animate.shift([0,1,0])
        )
        #self.wait()


        ############# First round PCR - anneal primer ########################
        
        a=Arrow(
            add_vec(pos_ins,[-1,dna_dy, 0]),
            add_vec(pos_ins,[+1,dna_dy, 0]),
            color="BLUE"
        )
        a.shift([0,10,0])
        self.add(a)
        self.play(
            a.animate.shift([0,-10,0])
        )
        self.wait()

        ############# First round PCR - extend primer ########################
        
        a2=Arrow(
            add_vec(pos_ins,[-1.0, dna_dy, 0]),
            add_vec(pos_ins,[+3.4, dna_dy, 0]),
            color="BLUE"
        )

        self.play(
            a.animate.become(a2)
        )
        self.wait()


        ############# Create text ########################
        
        self.play(
            txt_pcr.animate.shift([0,5,0])
        )
        txt_pcr2 = Tex(r"Next cycle targets \textit{new} complementary strand")
        txt_pcr2.move_to([0,3,0])
        self.play(
            Create(txt_pcr2)
        )
        
        ############# Second round PCR - "heat up" ########################
        
        group_upper_strand = VGroup([dna_ins[1]], spl_ins[0], spl_ins[2], spr_ins[1])
        self.play(
            group_upper_strand.animate.shift([0,-1,0])
        )
        self.wait()

        ############# Second round PCR - anneal primer ########################
        
        a=Arrow(
            add_vec(pos_ins,[+1, -dna_dy, 0]),
            add_vec(pos_ins,[-1, -dna_dy, 0]),
            color="GREEN"
        )
        a.shift([2.4, 10, 0])
        self.add(a)
        self.play(
            a.animate.shift([0,-10,0])
        )
        self.wait()



        ############# Second round PCR - extend primer ########################
        
        a2=Arrow(
            add_vec(pos_ins,[+1.0, -dna_dy, 0]),
            add_vec(pos_ins,[-3.4, -dna_dy, 0]),
            color="GREEN"
        )
        a2.shift([2.4, 0, 0])

        self.play(
            a.animate.become(a2)
        )
        self.wait()
        

        ############# Zoom in ########################

        self.play(
            self.camera.frame.animate.scale(0.5).move_to([-2,0,0])
        )
        self.wait()

        ############# Create text ########################
        
        txt_pcr3 = Tex(r"Final amplicon contains insert + surrounding DNA")
        txt_pcr3.move_to([-2,0.5,0])
        txt_pcr3.scale(0.5)
        self.play(
            Create(txt_pcr3)
        )
        self.wait()
```
